# API Local Testing Check
This notebook verifies that the FastAPI web server can initialize successfully, and we can connect to the `/ws/voice` WebSocket endpoint without crashing.

```python
import subprocess
import time
import os
import sys

# Start the FastAPI server using the existing venv
# Using the relative path assumes the notebook is run from `local_testing/`
cwd = os.path.abspath("../backend")
uvicorn_path = os.path.join(cwd, "venv", "bin", "uvicorn")

server_process = subprocess.Popen(
    [uvicorn_path, "main:app", "--host", "127.0.0.1", "--port", "8000"],
    cwd=cwd
)

print("Starting FastAPI server...")
time.sleep(3)  # Give the server a few seconds to start up
```

```python
import asyncio
import websockets

async def test_websocket():
    uri = "ws://127.0.0.1:8000/ws/voice?session_id=test_session_123"
    print(f"Connecting to {uri}...")
    try:
        async with websockets.connect(uri) as websocket:
            print("Successfully connected!")
            
            # We can immediately close the connection to see if it gracefully terminates.
            await websocket.close()
            print("Closed WebSocket connection.")
    except Exception as e:
        print(f"Failed to connect to /ws/voice endpoint. Error: {e}")

# Await inside jupyter functions normally as notebooks patch the event loop
await test_websocket()
```

```python
# Cleanup: terminate the server
server_process.terminate()
server_process.wait()
print("Server terminated successfully.")
```
